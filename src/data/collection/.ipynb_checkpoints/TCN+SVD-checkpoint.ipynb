## 1️⃣ 데이터 전처리
### 전치

````python
nutrition = pd.read_json("./nutrition_db.json")
nutrition_df = nutrition.T.reset_index(drop=True)
````

### 숫자형 변환

````python
nutrition_df["kcal"] = nutrition_df["kcal"].astype(float)

nutrition_df["protein"] = (
    nutrition_df["protein"]
    .str.replace("g", "", regex=False)
    .astype(float)

)

nutrition_df["carbs"] = (
    nutrition_df["carbs"]
    .str.replace("g", "", regex=False)
    .astype(float)

)
nutrition_df.head()
````

### ✅ 핵심 입력값

````python
user = {
    "weight": 70,           # kg
    "muscle_mass": 30,      # 골격근량
    "body_fat": 18,         # 체지방률 (%)
    "visceral_fat": 8,      # 내장지방
    "bmr": 1600,            # 기초대사량
    "activity": True,       # 운동 여부
    "goal": "bulk"          # bulk / diet / lean
}
````

## 🔥 2. 유지 칼로리 계산

````python
def calculate_maintenance_calories(user):
    activity_factor = 1.55 if user["activity"] else 1.2
    return user["bmr"] * activity_factor
````

## 🔥 3. 목표별 칼로리 설정

````python
def calculate_target_calories(user):
    maintenance = calculate_maintenance_calories(user)

    if user["goal"] == "bulk":
        return maintenance + 500
    elif user["goal"] == "diet":
        return maintenance - 500
    elif user["goal"] == "lean":
        return maintenance + 300
````

## 🔥 4. 탄단지 비율 설정

````python
def get_macro_ratio(goal):
    if goal == "bulk":
        return {"carbs": 0.5, "protein": 0.3, "fat": 0.2}
    elif goal == "diet":
        return {"carbs": 0.4, "protein": 0.4, "fat": 0.2}
    elif goal == "lean":
        return {"carbs": 0.45, "protein": 0.35, "fat": 0.2}
````

## 🔥 5. 단백질 필요량 계산 (핵심)

````python
def calculate_protein_requirement(user):
    if user["activity"]:
        return user["weight"] * 1.8  # 평균값
    else:
        return user["weight"] * 1.0
````

## 🔥 6. 음식 추천 로직

### 핵심 아이디어

```plaintext
✔ 단백질 높은 음식 우선
✔ 지방 낮은 음식 (다이어트)
✔ 칼로리 맞춤 필터링
```

### 데이터 전처리 (중요)

````python
nutrition_df["fat"] = nutrition_df["fat"].str.replace("g", "").astype(float)
````

---

### 추천 함수

````python
def recommend_foods(nutrition_df, user):
    protein_target = calculate_protein_requirement(user)
    target_calories = calculate_target_calories(user)

    if user["goal"] == "diet":
        # 고단백 저지방
        filtered = nutrition_df[
            (nutrition_df["protein"] > 15) &
            (nutrition_df["fat"] < 10)
        ]
    elif user["goal"] == "bulk":
        # 고칼로리 고단백
        filtered = nutrition_df[
            (nutrition_df["protein"] > 15) &
            (nutrition_df["kcal"] > 150)
        ]
    else:
        # 균형
        filtered = nutrition_df[
            (nutrition_df["protein"] > 10)
        ]

    # 단백질 기준 정렬
    return filtered.sort_values(by="protein", ascending=False).head(5)
````

---
### 🚀 실행 예시

````python
recommend_foods(nutrition_df, user)
````

 ## 4. SVD 적용

````python
ratings_df = pd.DataFrame({
    "user_id": [1, 1, 2, 2],
    "food_name": ["닭가슴살", "삼겹살", "새우", "소고기"],
    "rating": [5, 2, 4, 5],
    
    # 🔥 추가 추천 (고급)
    "goal": ["diet", "bulk", "diet", "bulk"],
    "timestamp": ["2026-05-01", "2026-05-02", "2026-05-03", "2026-05-04"]
})
ratings_df = ratings_df[["user_id", "food_name", "rating"]]
ratings_df
````

````python
nutrition_df.rename(columns={"name": "food_name"}, inplace=True)
nutrition_df["food_name"] = nutrition_df["food_name"].str.replace("불고기감 ", "")
nutrition_df.head()
````

### 식단 로그 생성 (핵심)

````python
diet_log = pd.DataFrame({
    "user_id": [1,1,1,1],
    "date": ["2026-05-01","2026-05-02","2026-05-03","2026-05-04"],
    "food_name": ["닭가슴살","삼겹살","새우","소고기"]
})

# 영양성분 붙이기
diet_log = diet_log.merge(nutrition_df, on="food_name")
````

---

### ✅ SVD 추천 모델

````python
from surprise import Dataset, Reader, SVD

reader = Reader(rating_scale=(1, 5))
data = Dataset.load_from_df(ratings_df, reader)
trainset = data.build_full_trainset()

svd_model = SVD()
svd_model.fit(trainset)
````

### 🔥 추천 점수 생성

### ✅️ 시계열 데이터 생성 (TCN용)
#### 하루 총 칼로리

````python
diet_log["date"] = pd.to_datetime(diet_log["date"])

daily_kcal = diet_log.groupby("date")["kcal"].sum()
````

#### darts 변환

````python
from darts import TimeSeries
import numpy as np

ts = TimeSeries.from_series(daily_kcal.astype(np.float32))
````

### ✅️ TCN 모델

````python
from darts.models import TCNModel
from darts.dataprocessing.transformers import Scaler

scaler = Scaler()
ts_scaled = scaler.fit_transform(ts)

model = TCNModel(
    input_chunk_length=3,
    output_chunk_length=1,
    n_epochs=50,
    random_state=0,
    kernel_size=2 
)

model.fit(ts_scaled)
````

### 🔥 미래 칼로리 예측

````python
pred = model.predict(3)  # 3일 예측

pred_kcal = scaler.inverse_transform(pred).values().flatten()
future_kcal = pred_kcal.mean()
````

### ✅ ️ 추천 보정 (핵심 로직🔥)
#### 사용자 목표 설정

````python
target_kcal = 2000  # 예시
target_protein = 120
````

#### 영양성분 + SVD 결합

##### nutrition_df 기준 전체 음식 예측

````python
predictions = []

for food in nutrition_df["food_name"]:

    pred = svd_model.predict(
        uid=1,
        iid=food
    )

    predictions.append({
        "food_name": food,
        "svd_score": pred.est
    })

svd_df = pd.DataFrame(predictions)

svd_df.head()
````

````python
final_df = svd_df.merge(nutrition_df, on="food_name")
final_df.head()
````

#### 점수 계산

````python
goal = "diet"   # 또는 "bulk"

def score(row):
    score = row["svd_score"]
    # -------------------
    # 다이어트
    # -------------------
    if goal == "diet":
        score += row["protein"] * 0.06
        score -= row["fat"] * 0.05
        score -= row["carbs"] * 0.03
        score -= row["kcal"] * 0.003

    # -------------------
    # 벌크업
    # -------------------
    elif goal == "bulk":
        score += row["protein"] * 0.05
        score += row["carbs"] * 0.04
        score -= row["fat"] * 0.01

    # -------------------
    # 칼로리 초과 패널티
    # -------------------
    if future_kcal + row["kcal"] > target_kcal:
        score -= 2

    return score
````

#### 추천 결과

````python
recommend = final_df.sort_values(
    by="final_score",
    ascending=False
)

recommend[[
    "food_name",
    "svd_score",
    "protein",
    "fat",
    "final_score"
]].head(10)
````

