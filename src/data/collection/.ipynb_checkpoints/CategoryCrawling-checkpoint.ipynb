```python
# Crawling Tool
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait # 대기할 때 필요
from selenium.webdriver.support import expected_conditions as EC # 상태 확인 시 필요
# Parsing Library
from bs4 import BeautifulSoup

import time
```

```python
# driver 셋팅 함수 
def get_safe_driver():
    chrome_options = Options()
    chrome_options.add_argument("--headless") # 실제 구동 시에는 Headless 권장
    chrome_options.add_argument("--no-sandbox")
    chrome_options.add_argument("--disable-dev-shm-usage")
    chrome_options.add_argument("--disable-gpu")
    chrome_options.add_argument("--blink-settings=imagesEnabled=false")
    chrome_options.page_load_strategy = 'eager'
    chrome_options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36")
    
    driver = webdriver.Chrome(options=chrome_options)
    return driver
```

```python
# 페이지 파싱 함수
def PageParsing(driver):
    html = driver.page_source
    soup = BeautifulSoup(html, "html.parser")
    return soup
```

```python
url = "https://www.lampcook.com/cook/food_top100_list5.php"
work_driver = get_safe_driver()
work_driver.get(url)
```

```python
try:
    # button selector
    button_selector = "#div_btn_span_txt"
    work_driver.find_element(By.ID, "div_btn_span_txt").click()
    
    time.sleep(2)
    
    soup = PageParsing(work_driver)
    
    food_names = soup.select("#shell_data_div1 > div.def_color_box > div")
    food_names2 = soup.select("#shell_data_div2 > div.def_color_box > div") 

    food_imgs = soup.selec("#shell_data_div1 > div > div > img")
    food_imgs2 = soup.selec("#shell_data_div2 > div > div > img")
    
    food_name = [name.text for name in food_names]
    food_name2 = [name.text for name in food_names2 ] 
    name_list = food_name + food_name2
    

    food_img = [img['src'] for img in food_imgs] 
    
    print(name_list)


except Exception as e:
    print(f"error: {e}")

finally: 
    work_driver.quit()
```
