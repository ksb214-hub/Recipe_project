# AI 추천 시스템 설계 — 협업 필터링(Collaborative Filtering)

---

## 추천 시스템 구조

```mermaid
flowchart TD

A[사용자 행동 데이터 수집]

A --> B[사용자-레시피 행렬 생성]

B --> C[SVD 행렬 분해]

C --> D[유사도 계산]

D --> E[추천 점수 예측]

E --> F[개인화 음식 추천]
```

---

## 왜 SVD인가?
### 장점
- 구현 쉬움
- 추천 정확도 높음
- 데이터 적어도 가능
- 음식 추천과 궁합 좋음

## 1️⃣ 데이터 준비

````python
import pandas as pd

ratings = pd.DataFrame({
    "user_id": [1,1,2,2,3,3],
    "recipe_id": [
        "불고기",
        "파스타",
        "불고기",
        "샐러드",
        "파스타",
        "샐러드"
    ],
    "rating": [5,4,5,1,4,5]
})
ratings
````

## 2️⃣ SVD 모델

````python
from surprise import Dataset
from surprise import Reader
from surprise import SVD
from surprise.model_selection import train_test_split

reader = Reader(rating_scale=(1,5))

data = Dataset.load_from_df(
    ratings[["user_id","recipe_id","rating"]],
    reader
)

trainset, testset = train_test_split(
    data,
    test_size=0.2
)

model = SVD()

model.fit(trainset)
````

## 3️⃣ 추천 예측
### 점수 기준

```python
Reader(rating_scale=(1,5))
```

로 설정

---

| 점수 | 의미 |
|-----|-----|
| 1 | 매우 싫어함 |
| 2 | 별로 |
| 3 | 보통 |
| 4 | 좋아함 |
| 5 | 매우 좋아함 |


````python
prediction = model.predict(
    uid=1,
    iid="샐러드"
)
prediction.est.round(2)
````
