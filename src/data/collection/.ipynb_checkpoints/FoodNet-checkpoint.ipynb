````python
import torch
from torchvision import datasets, models, transforms
import matplotlib.pyplot as plt
import os
import random
import ssl
import certifi
import os
from torch.utils.data import DataLoader
import torch.nn as nn
import numpy as np
import torch.optim as optim
from PIL import ImageFile
````

````python
ImageFile.LOAD_TRUNCATED_IMAGES = True
base_path = "/Users/gangseongbin/login-project/src/data/collection/data"
for root, dirs, files in os.walk(base_path):
    print(root)
````

## 2. MPS(GPU) 장치 설정 및 모델 생성 
###  왜 ResNet 쓰냐?

* 이미지 분류에서 검증된 모델
* Food101에 매우 적합
* Transfer Learning 가능 (핵심)

````python
ssl._create_default_https_context = lambda: ssl.create_default_context(
    cafile=certifi.where()
)

device = torch.device(
    "mps" if torch.backends.mps.is_available()
    else "cpu"

)

print("현재 사용 장치:", device)
# 모델을 GPU로 보내기
model = models.resnet101(weights=None)
model.fc = nn.Linear(model.fc.in_features, 101)
model.to(device)
````

## 3. 데이터 전처리(Transform)

````python
# 이미지를 Tensor 형태로 변환

transform = transforms.Compose([
    # CNN 입력 크기 통일
    transforms.Resize((224, 224)),
    # Tensor 변환
    transforms.ToTensor(),
    # 정규화
    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )

])
````

## 4. Food101 데이터셋 다운로드 및 로드
```root
- 데이터가 저장될 폴더
- split='train'
- 학습용 데이터 사용
```

````python
food_dataset = datasets.Food101(
    root="/Users/gangseongbin/login-project/src/data/collection/data",
    split="train",
    download=False,
    transform=transform
)
````

## 5. Food101 클래스(카테고리) 확인
> Food101은 총 101개의 음식 카테고리를 보유

````python
classes = food_dataset.classes
print("총 음식 클래스 개수:")
print(len(classes))
print("\n일부 음식 카테고리 확인:")
print(classes[:20])
````

### 5-1. train_dataset 생성

````python
train_dataset = datasets.Food101(
    root="/Users/gangseongbin/login-project/src/data/collection/data",
    split="train",
    download=False,   # 이미 다운로드 했으니까 False
    transform=transform

)
````

### 5-2. test_dataset 생성

````python
test_dataset = datasets.Food101(
    root="/Users/gangseongbin/login-project/src/data/collection/data",
    split="test",
    download=False,
    transform=transform
)
````

## 6. DataLoader 생성 

````python
train_loader = DataLoader(
    train_dataset,
    batch_size=32,
    shuffle=True,
    num_workers=0  # 🔥 Mac 필수

)

test_loader = DataLoader(
    test_dataset,
    batch_size=32,
    shuffle=False,
    pin_memory=True
        
)
````

## 7. 손실 함수 + 옵티마이저

````python
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(
    model.parameters(),
    lr=0.001
)
````

## 8. 학습 루프

````python
for i in range(len(train_dataset)):
    try:
        img, label = train_dataset[i]
    except Exception as e:
        print(f"❌ 문제 이미지 index: {i}")
        print(e)
        break
````

````python
epochs = 5

for epoch in range(epochs):
    model.train()
    running_loss = 0.0

    for images, labels in train_loader:
        images = images.to(device)
        labels = labels.to(device)

        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)

        loss.backward()
        optimizer.step()

        running_loss += loss.item()
        
    print(f"Epoch {epoch+1}, Loss: {running_loss/len(train_loader)}")
````

## 9. 모델 평가

```plaintext
1. model.eval() → 평가 모드 전환
2. gradient 끔 → 속도/메모리 최적화
3. 예측값 vs 정답 비교
4. 전체 대비 맞춘 비율 계산
```

````python
# 1️⃣ 평가 모드
# 👉 Dropout / BatchNorm 비활성화
model.eval()

correct = 0
total = 0

# 2️⃣ gradient 비활성화
# 👉 학습이 아니라 평가니까 필요 없음
# 👉 속도 ↑ 메모리 ↓
with torch.no_grad():
    for images, labels in test_loader:

        images = images.to(device)
        labels = labels.to(device)
        # 3️⃣ 예측
        # 👉 가장 확률 높은 클래스 선택
        outputs = model(images)
        _, predicted = torch.max(outputs, 1)
        
        # 4️⃣ 정확도 계산
        # 👉 맞춘 개수 누적
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

# 👉 최종 정확도 (%)
accuracy = 100 * correct / total

print(f"Accuracy: {accuracy:.2f}%")
````

## 10. 예측 테스트 (직접 확인)

````python
classes = train_dataset.classes

def predict_image(image, model):
    model.eval()
    
    image = image.unsqueeze(0).to(device)

    with torch.no_grad():
        outputs = model(image)
        _, predicted = torch.max(outputs, 1)

    return classes[predicted.item()]
````

## 📊 결과 시각화

````python
images, labels = next(iter(test_loader))

img = images[0]
label = labels[0]
pred = predict_image(img, model)

plt.imshow(img.permute(1,2,0))
plt.title(f"True: {classes[label]} / Pred: {pred}")
plt.axis('off')
plt.show()
````
